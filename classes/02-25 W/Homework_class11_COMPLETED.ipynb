## Problem 1 — `is_prime(n)`

Write a function `is_prime(n)` that returns whether `n` is a prime number.

**Assumptions**
- `n` is a **positive integer**.
- A prime number is an integer greater than 1 that has exactly two divisors: 1 and itself.

**Examples**
- `is_prime(2) -> True`
- `is_prime(15) -> False`

**Notes**
- Your solution should be correct for small and moderately large values of `n`.
- You may choose any reasonable approach (trial division, checking up to √n, etc.).


````python
def is_prime(n: int) -> bool:
    """
    Return True if n is prime.
    Return False otherwise.
    Assume n is a positive integer.
    Example:
    is_prime(2) -> True
    is_prime(15) -> False
    """
    # edge cases first
    if n < 2:
        return False
    if n == 2:
        return True
    # even numbers bigger than 2 arent prime
    if n % 2 == 0:
        return False
    
    # only need to check up to sqrt(n)
    # if n has a divisor bigger than sqrt(n), it also has one smaller
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2  # skip even numbers
    
    return True

# quick tests
print(is_prime(2))   # True
print(is_prime(15))  # False  
print(is_prime(17))  # True
print(is_prime(1))   # False
````

## Problem 3 — `largest_product_of_6(s)`

Given a **string of digits** `s` (length ≥ 6), return the **largest product** of any **6 consecutive digits**.

**Example**
- `"1234567" -> 5040` (because `2*3*4*5*6*7 = 5040` is the largest product among all 6-digit windows)

**Assumptions**
- `s` contains only the characters `'0'` through `'9'`.
- `len(s) >= 6`

### How to enter the big digit block (copy/paste) and clean it

You'll often be given a large block of digits formatted across many lines (sometimes with spaces).  
To use it for computation, you need a **single continuous string** containing only digits (no spaces, no newlines).

Here are several common ways to do that in Python:

#### 
You can paste the digits into a triple-quoted string:

```python
digits_raw = '''

9698352031...
...
'''
```

This preserves newlines inside the string.

#### Option A — Remove only newline characters
If you only want to remove `\n` line breaks:

```python
digits_no_newlines = digits_raw.replace("\n", "")
```

This removes newlines but **does not** remove spaces or tabs.

#### Option B — Remove *all* whitespace (recommended / safest)
If there might be spaces, tabs, or newlines, remove all whitespace:

```python
digits_clean = ''.join(digits_raw.split())
```

- `split()` without arguments splits on **any whitespace**
- `join(...)` stitches everything back together with **no separators**

**Goal:** produce `digits_clean` that looks like `"7316717653...63450"` with no whitespace.

Once you have the cleaned string, your `largest_product_of_6(s)` function should work on it.


````python
def largest_product_of_6(s: str) -> int:
    """
    Given a string of digits (length >= 6),
    return the largest product of any 6 consecutive digits.
    Example:
    "1234567" -> 5040
    Assume s contains only digits 0-9.
    """
    # first clean up the string - get rid of whitespace
    s = ''.join(s.split())
    
    max_product = 0
    
    # slide a window of 6 across the string
    for i in range(len(s) - 5):
        # grab 6 consecutive digits
        window = s[i:i+6]
        
        # multiply them together
        product = 1
        for char in window:
            product *= int(char)
        
        if product > max_product:
            max_product = product
    
    return max_product

# test with the example
print(largest_product_of_6("1234567"))  # should be 5040

digits_raw = '''
    73167176 531330624919225119674426574742355349194934
    96983520312774506326239578318016984801869478851843
    85861560789112949495459501737958331952853208805511
    12540698747158523863050715693290963295227443043557
    66896648950445244523161731856403098711121722383113
    62229893423380308135336 276614282806444486645238749
    30358907296290491560440772390713810515859307960866
    70172427121883998797908792274921901699720888093776
    65727333001053367881220235421809751254540594752243
    52584907711670556013604839586446706324415722155397
    5369781797784617406495514929 0862569321978468622482
    83972241375657056057490261407972968652414535100474
    82166370484403199890008895243450658541227588666881
    16427171479924442928230863465674813919123162824586
    17866458359 124566529476545682848912883142607690042
    24219022671055626321111109370544217506941658960408
    07198403850962455444362981230987879927244284909188
    84580156166097919133875499200524063689912560717606
    05886116467109405077541002256983155200055935729725
    71636269561882670428252483600823257530420752963450
    '''

result = largest_product_of_6(digits_raw)
print(f"Largest product of 6 consecutive digits: {result}")
````

## Problem 4 — `word_histogram(text)`

Given a string of text, return a dictionary mapping each word to its frequency.

**Assumptions / Rules**
- Words are separated by whitespace.
- Convert words to lowercase.
- Ignore punctuation.

**Example**
- `"apple banana apple" -> {'apple': 2, 'banana': 1}`

**Notes**
- "Ignore punctuation" means characters like `. , ! ? : ; ' " ( )` should not be part of the words.
- You can choose any reasonable approach:
  - replace punctuation with spaces, then split
  - remove punctuation characters, then split
  - use `string.punctuation` if you want


````python
import string

def word_histogram(text: str) -> dict[str, int]:
    """
    Given a string of text, return a dictionary
    mapping each word to its frequency.
    Assume:
    - Words are separated by whitespace.
    - Convert words to lowercase.
    - Ignore punctuation.
    Example:
    "apple banana apple" -> {'apple': 2, 'banana': 1}
    """
    # lowercase everything first
    text = text.lower()
    
    # remove punctuation - replace each punct char with empty string
    for char in string.punctuation:
        text = text.replace(char, '')
    
    # split into words
    words = text.split()
    
    # count em up
    histogram = {}
    for word in words:
        if word in histogram:
            histogram[word] += 1
        else:
            histogram[word] = 1
    
    return histogram

# tests
print(word_histogram("apple banana apple"))
# {'apple': 2, 'banana': 1}

print(word_histogram("Hello, world! Hello again."))
# {'hello': 2, 'world': 1, 'again': 1}
````
