# f-Strings: The Part After `:` (Format Spec)  
This notebook teaches a **small, practical subset** of Python’s f-string *format specification mini-language* — the part that comes after the `:`.

We’ll focus on a simple **slot order** you can memorize:

```
f"{value: [fill][align][width][.precision][type] }"
```

## The 4 slots (in order)
1. **Fill + Align** *(optional)*  
2. **Width** *(optional; minimum field width)*  
3. **.Precision** *(optional)*  
4. **Type** *(optional)*  

### Align
- `<` left
- `>` right
- `^` center

### Fill
- Any single character, but it only works **when align is present**
- Examples: `0>`, `.^`, `-<`

### Width
- A number like `10` meaning **minimum** width (pads if needed)

### Precision
- `.2` is common with floats → `.2f` means 2 digits after the decimal

### Type (common)
- `d` integer
- `f` fixed-point float
- `%` percent

## Golden examples
```python
f"{n:>5}"       # right-align, width 5
f"{n:0>5}"      # zero-fill, right-align, width 5
f"{x:.2f}"      # 2 decimals
f"{x:8.2f}"     # width 8, 2 decimals
f"{p:.1%}"      # percent, 1 decimal
f"{name:^10}"   # center, width 10
```


````python
# Quick demo cell (run this)
n = 42
x = 3.14159
p = 0.2375
name = "Ava"

print(f"{n:>5}  <- right align width 5")
print(f"{n:0>5}  <- zero pad width 5")
print(f"{x:.2f}  <- 2 decimals")
print(f"{x:8.2f}  <- width 8, 2 decimals")
print(f"{p:.1%}  <- percent, 1 decimal")
print(f"{name:^10}<- centered width 10")
````

## One key idea
**Width is a minimum.** It pads if the value is shorter, but it never truncates.

Example:
```python
print(f"{123456:5}")  # prints 123456 (does not cut it off)
```


````python
# Width is minimum (run this)
print(f"{42:5}")       # padded to width 5
print(f"{123456:5}")   # expands beyond width 5
````
