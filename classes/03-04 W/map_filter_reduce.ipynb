## Setup
- In Python 3, `map()` and `filter()` return **iterators**, so you often wrap them in `list(...)` to see results.
- `reduce` lives in `functools`.


```python
from functools import reduce

def show(label, value):
    print(f"{label}: {value}")
```

## What is `lambda`?
`lambda` is Python’s way to create a **small anonymous function** (a function with **no name**) in one expression.

Example:
- Named: `def square(x): return x*x`
- Lambda: `lambda x: x*x`

### Why use it?
`lambda` is useful when:
- The function is **tiny** (1 expression)
- You only need it **once**
- You want to pass a function *as an argument* (like to `map`, `filter`, `reduce`, `sorted(key=...)`)

### When NOT to use it
If it takes more than a line to understand, **use a named function**.


```python
def square(x):
    return x * x

square_lambda = lambda x: x * x

show("square(5)", square(5))
show("square_lambda(5)", square_lambda(5))
```

## 1) `map(function, iterable)`
**Purpose:** transform each element.

If `items = [a, b, c]`, then `map(f, items)` behaves like:
- `[f(a), f(b), f(c)]` (but lazily)

### First: with a named function


```python
def c_to_f(c):
    return c * 9/5 + 32

celsius = [0, 10, 20, 30]
fahrenheit = list(map(c_to_f, celsius))

show("celsius", celsius)
show("fahrenheit", fahrenheit)
```

### Same thing: with `lambda`
This is handy when the transformation is short and used once.


```python
celsius = [0, 10, 20, 30]
fahrenheit = list(map(lambda c: c * 9/5 + 32, celsius))
show("fahrenheit", fahrenheit)
```

### `map` with multiple iterables
`map` can take multiple iterables. The function must accept that many arguments.
It stops at the shortest iterable.

Named first, then lambda.


```python
def add(x, y):
    return x + y

a = [1, 2, 3]
b = [10, 20, 30, 40]

show("map with named", list(map(add, a, b)))
show("map with lambda", list(map(lambda x, y: x + y, a, b)))
```

## 2) `filter(predicate, iterable)`
**Purpose:** keep elements that pass a test.

If `items = [a, b, c]`, then `filter(p, items)` behaves like:
- `[x for x in items if p(x)]` (but lazily)

### First: with a named predicate function


```python
def is_even(x):
    return x % 2 == 0

nums = list(range(-5, 6))
evens = list(filter(is_even, nums))
show("nums", nums)
show("evens", evens)
```

### Same thing: with `lambda`


```python
nums = list(range(-5, 6))
evens = list(filter(lambda x: x % 2 == 0, nums))
show("evens", evens)
```

### `filter(None, iterable)`
Special case: if the predicate is `None`, `filter` keeps only **truthy** values.


```python
items = ["", "hi", None, "bye", 0, 3, [], [1,2]]
truthy_only = list(filter(None, items))
show("truthy_only", truthy_only)
```

## 3) `reduce(function, iterable[, initializer])`
**Purpose:** combine many values into one.

It repeatedly updates an **accumulator**:
- `acc = f(acc, next_item)`

### First: with a named function


```python
def add(acc, x):
    return acc + x

nums = [1, 2, 3, 4]
sum_via_reduce = reduce(add, nums)  # no initializer -> starts with first item
show("sum_via_reduce", sum_via_reduce)
```

### Same thing: with `lambda`


```python
nums = [1, 2, 3, 4]
sum_via_reduce = reduce(lambda acc, x: acc + x, nums)
show("sum_via_reduce", sum_via_reduce)
```

### Using an initializer
Initializer sets the starting accumulator.
Example: product starting at 1.


```python
def multiply(acc, x):
    return acc * x

nums = [1, 2, 3, 4]
prod = reduce(multiply, nums, 1)
show("prod", prod)
```
